```python
#1.Obtén o nome de todos os equipos da Liga Española da tempada 2015-2016
```

```python
url_base="https://raw.githubusercontent.com/openfootball/football.json/master/"
busqueda_espana="2015-16/es.1.clubs.json"
url_equipos_espana=url_base+busqueda_espana
```

```python
response = requests.get(url_equipos_espana)
resposta=response.json()
equipos=resposta['clubs']
lista_equipos=[]
for i in range(0,len(equipos)):
    lista_equipos.append(equipos[i]['name'])
    print(lista_equipos[i])
    i+1;
```

```python
#2.Obtén os resultados da Xornada 1 da Liga Española da tempada 2012-2013
```

```python
busqueda_espana="2012-13/es.1.json"
url_resultado_espana=url_base+busqueda_espana
```

```python
response = requests.get(url_resultado_espana)
resposta=response.json()
lista_resultados=[]
lista_locales=[]
lista_visitantes=[]
for i in range(0,10):
    lista_locales.append(resposta['rounds'][0]['matches'][i]['team1'])
    lista_visitantes.append(resposta['rounds'][0]['matches'][i]['team2'])
    lista_resultados.append(resposta['rounds'][0]['matches'][i]['score']['ft'])
    print("Na xornada 1 da tempada 2012-2013, enfrontaronse:",lista_locales[i],"-",lista_visitantes[i],"e o resultado foi: ",lista_resultados[i])
    i+1;
```

```python
#3.Mostra os equipos dos grupos [A,B,D e F] da Champions de 2013-2014
```

```python
busqueda_ucl="2013-14/cl.groups.json"
url_grupos_ucl=url_base+busqueda_ucl
```

```python
response=requests.get(url_grupos_ucl)
resposta=response.json()
grupo_a=resposta['groups'][0]['teams']
grupo_b=resposta['groups'][1]['teams']
grupo_d=resposta['groups'][3]['teams']
grupo_f=resposta['groups'][5]['teams']
print("Os equipos do grupo A ese ano foron : {}".format(grupo_a))
print("Os equipos do grupo B ese ano foron : {}".format(grupo_b))
print("Os equipos do grupo D ese ano foron : {}".format(grupo_d))
print("Os equipos do grupo F ese ano foron : {}".format(grupo_f))
```

```python
#4.Mostra que equipos disputaron a final de Champions de 2016 e cal foi o resultado
```

```python
busqueda_final_ucl="2015-16/cl.json"
url_final_ucl=url_base+busqueda_final_ucl
```

```python
response=requests.get(url_final_ucl)
resposta=response.json()
equipo1=resposta['rounds'][-1]['matches'][0]['team1']
equipo2=resposta['rounds'][-1]['matches'][0]['team2']
resultado=resposta['rounds'][-1]['matches'][0]['score']['ft']
print("Na tempada 2015-2016, a final da UCL foi {} - {} e o resultado final foi {}".format(equipo1,equipo2,resultado))
```
